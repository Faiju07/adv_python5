```python
class Person :
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def introduction(self):
        print(f'Hi there i am {self.name}')
        print(f'i am {self.age} years old')

class Employee(Person):
    def __init__(self, name, age, Company,salary):
        super().__init__(name, age)
        self.Company = Company
        self.salary = salary

    def work(self):
        super().introduction()
        print(f'I work for {self.Company}.')   

    def salary_hike(self):
        super().introduction()
        print('My Current Salary is', self.salary, 'and after hike is', round(self.salary*1.10), 'RS')
```

#### Multilevel Inheritence

```python
class Employees:
    def __init__(self, name:str,emp_id:int):
        self.name = name
        self.emp_id = emp_id

    def get_employees_info(self):
        print( f'Name is {self.name} and Employee id is {self.emp_id}')
    
class Manager(Employees):
    def __init__(self, name:str, emp_id:int, department):
        super().__init__(name, emp_id)
        self.department = department

    def get_manager_info(self):
        super().get_employees_info()
        print( f'Department : {self.department}')
    
class ProjectManager(Manager):
    def __init__(self, name, emp_id, department,Project):
        super().__init__(name, emp_id, department)
        self.Project =Project

    def get_project_manager_info(self):
        super().get_manager_info()
        print( f'Project Manager :  {self.Project}')
```

```python
class Employees:
    def __init__(self, name:str,emp_id:int):
        self.name = name
        self.emp_id = emp_id

    def get_employees_info(self):
        return f'Name is {self.name} and Employee id is {self.emp_id}'
    
class Manager(Employees):
    def __init__(self, name:str, emp_id:int, department):
        super().__init__(name, emp_id)
        self.department = department

    def get_manager_info(self):
        super().get_employees_info()
        return f'{self.get_employees_info()}, Department : {self.department}'
    
class ProjectManager(Manager):
    def __init__(self, name, emp_id, department,Project):
        super().__init__(name, emp_id, department)
        self.Project =Project

    def get_project_manager_info(self):
        super().get_manager_info()
        return f'{self.get_manager_info()}, Project Manager :  {self.Project}'
```

```python
from dataclasses import dataclass
```

```python
@dataclass
class User:
    name : str
    email : str
    def get_user_info(self):
        print('Name is', self.name)
        print('Email is', self.email)

class Admin(User):
    def __init__(self, name, email, access : str):
        super().__init__(name, email)
        self.access = access

    def get_admin_info(self):
        super().get_user_info()
        print(f'Access:{self.access}')

class RegularUser(User):
    def __init__(self, name, email, subscription : bool):
        super().__init__(name, email)
        self.subscription = subscription

    def get_RegularUser_info(self):
        super().get_user_info()
        print(f'Subscription: {self.subscription}')
```

```python
u1 = User(name = 'Aman', email = 'amanemail.com')
```

```python
u1.name
```

```python
u1.email
```

```python

u1.get_user_info()
```

```python

from pydantic import BaseModel, Field, EmailStr
from datetime import date
from typing import List
```

```python
class Student(BaseModel):
    Name : str = Field(description= 'Enter name here:')
    Roll_no : int = Field(description= 'Enter roll_no here:', gt = 0)
    Email : EmailStr = Field(description= 'Enter email here:')
    Marks : float = Field(description= 'Enter marks here:')
    Doj : date = Field(description='Enter date of joining here:')
    Hobbies : List[str] = Field(description= 'List of Hobbies:', default_factory=list)
    
    def get_student_info(self):
        print('Name of student is', self.Name)
        print('Roll No of student:', self.Roll_no)
        print('Email of student is', self.Email)
        print('Marks of student are', self.Marks)
        print('Date of joining of student is:', self.Doj)
        print('Hobbies of student are', self.Hobbies)
```

```python
s1 = Student(Name = 'Anmol',
             Roll_no = 55,
             Email = 'Anmol34@gmail.com',
             Marks = 89.5,
             Doj = date(2020 , 3 ,4),
             Hobbies = ['Reading, Dancing'])
```

```python
s1.get_student_info()
```
